```python
path_test = glob('./Eye_Data/test/*/*')
len(path_test)
```

```python
DF_test = pd.DataFrame(path_test)
DF_test
```

```python
def imread(path):
    ans = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
    ans = cv2.resize(ans, (512,512), interpolation = cv2.INTER_AREA)
    return ans
```

```python
ans = DF_test[0].map(imread)
```

```python
ans = np.array(ans)
```

```python
z = []
for e in ans:
    z.append(e)
```
