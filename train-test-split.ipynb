```python
import re
from glob import glob
import shutil
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
paths = glob('data/*')
paths
```

```python
def label_maker(path):
    return re.split('[\\\./_]' ,path)[1]
```

```python
labels = paths.map(label_maker)
```

```python
DF = pd.DataFrame(paths, columns=['paths'])
```

```python
labels = DF['paths'].map(label_maker)
```

```python
labels
```

```python
DF
```

```python
xx_train, xx_test, yy_train, yy_test = train_test_split(paths, labels, test_size=0.15, stratify=labels)
# xx_train.shape, xx_test.shape, yy_train.shape, yy_test.shape
```

```python
len(yy_train[yy_train=='normal'])/len(yy_train), len(yy_test[yy_test=='normal'])/len(yy_test)
```

```python
for label,path in zip(yy_train, xx_train):
    if label == 'normal':
        shutil.move(path, 'train/0')
    if label == 'cataract':
        shutil.move(path, 'train/1')
```

```python
len(yy_train[yy_train=='normal']), len(yy_train[yy_train=='cataract'])
```

```python
for label,path in zip(yy_test, xx_test):
    if label == 'normal':
        shutil.move(path, 'test/0')
    if label == 'cataract':
        shutil.move(path, 'test/1')
```

```python
len(yy_test[yy_test=='normal']), len(yy_test[yy_test=='cataract'])
```
